## ğŸ“ˆ Matplotlib (`matplotlib.pyplot`)

**Matplotlib** es la biblioteca de visualizaciÃ³n fundamental en Python. `pyplot` es su mÃ³dulo principal que proporciona una interfaz similar a MATLAB para crear grÃ¡ficos.

---

### ğŸ¨ ConfiguraciÃ³n de figuras

#### Estructura bÃ¡sica
- **`plt.figure(figsize=(width, height))`** - Define el tamaÃ±o del grÃ¡fico en pulgadas
- **`plt.subplots(nrows, ncols, figsize)`** - Crea mÃºltiples subgrÃ¡ficos
- **`plt.tight_layout()`** - Ajusta automÃ¡ticamente el espaciado

#### TÃ­tulos y etiquetas
- **`plt.title(label, fontsize, color, loc)`** - AÃ±ade tÃ­tulo al grÃ¡fico
  - `loc`: `'left'`, `'center'`, `'right'`
- **`plt.xlabel(label, fontsize)`** - Etiqueta del eje X
- **`plt.ylabel(label, fontsize)`** - Etiqueta del eje Y
- **`plt.suptitle(title)`** - TÃ­tulo principal para mÃºltiples subgrÃ¡ficos

#### Leyendas
- **`plt.legend(loc, fontsize, frameon, shadow)`** - AÃ±ade leyenda
  - `loc`: `'best'`, `'upper right'`, `'lower left'`, etc.

---

### ğŸ“‰ Tipos de grÃ¡ficos bÃ¡sicos

#### ğŸ“ GrÃ¡fico de lÃ­neas
```python
plt.plot(x, y, color, linewidth, linestyle, marker, label)
```
- `color`: `'r'`, `'blue'`, `'#FF5733'`
- `linestyle`: `'-'`, `'--'`, `'-.'`, `':'`
- `marker`: `'o'`, `'s'`, `'^'`, `'*'`, `'D'`
- `linewidth`: Grosor de la lÃ­nea (nÃºmero)

#### ğŸ”µ GrÃ¡fico de dispersiÃ³n
```python
plt.scatter(x, y, c, s, alpha, marker, edgecolors)
```
- `s`: TamaÃ±o de los puntos
- `c`: Color o array de colores
- `alpha`: Transparencia (0-1)

#### ğŸ“Š GrÃ¡fico de barras
```python
plt.bar(x, height, width, color, alpha, edgecolor)
plt.barh(y, width)  # Barras horizontales
```
- `width`: Ancho de las barras (0-1)
- `edgecolor`: Color del borde

#### ğŸ“ˆ Histograma
```python
plt.hist(x, bins, color, alpha, edgecolor, density)
```
- `bins`: NÃºmero de intervalos o lista de lÃ­mites
- `density`: `True` para normalizar

#### ğŸ¥§ GrÃ¡fico circular
```python
plt.pie(x, labels, autopct, colors, explode, startangle)
```
- `autopct`: Formato de porcentajes (`'%1.1f%%'`)
- `explode`: SeparaciÃ³n de sectores
- `startangle`: Ãngulo de inicio (grados)

### ğŸ’¾ Control de visualizaciÃ³n

- **`plt.show()`** - Muestra el grÃ¡fico
- **`plt.savefig(filename, dpi, bbox_inches, transparent)`** - Guarda el grÃ¡fico
  - `dpi`: ResoluciÃ³n (ej: 300 para alta calidad)
  - `bbox_inches='tight'`: Elimina espacios en blanco
  - Formatos: `.png`, `.jpg`, `.pdf`, `.svg`
