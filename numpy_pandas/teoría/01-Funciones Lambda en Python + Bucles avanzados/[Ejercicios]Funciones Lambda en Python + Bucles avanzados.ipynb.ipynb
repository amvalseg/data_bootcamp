# Ejercicio 2
Escriba una función de Python para dar la suma de los números de una lista.

````python
def suma(lista_numeros):
  """Funcion que realiza la suma de los elementos de una lista """
  suma = 0
  for numero in lista_numeros:
    suma = suma + numero
  return suma
 
lista_numeros = [11, 6, 1, 4]
 
print(suma(lista_numeros))
````

# Ejercicio 3

Dada una lista y un número, cree una función que devuelva dos parámetros:

1) Si está dicho número en esa lista. En dicho caso Devolverá `TRUE`. En caso contrario, `False`.

2) Si la longitud de la lista coincide con dicho número. En dicho caso devolverá `TRUE`. En caso contrario, `False`.

````python
def doble_salida(lista_numeros, numero_comprobar):
  esta_num_presente = True if numero_comprobar in lista_numeros else False
  longitud_igual = True if len(lista_numeros) == numero_comprobar else False

  return esta_num_presente, longitud_igual
 
print(doble_salida([1,2,3,4,5], 5))
````

#Ejercicio 7
Cree una función para contar la cantidad de números negativos de una lista.


````python
def contar_negativos(lista_numeros):
    if not lista_numeros: return []
    return [len([numero for numero in lista_numeros if numero < 0])]
lista = [22, 5, -76, 100, -12.3, 9]
print(contar_negativos(lista))
````

#Ejercicio 9

Dada una palabra, cuente el número de vocales que tiene.



````python
def check_vocales(palabra):
  vocales = ['a', 'e', 'i', 'o', 'u']
  total_vocales = 0
  for vocal in vocales:
    if vocal in palabra:
      total_vocales += 1
  return total_vocales

print(check_vocales("Hola"))
````


**3**) Encuentre aquellos factores superiores a dicho número. Para localizar un elemento de la matrix puede usar:

```

```

**4**) Devuelva su matrix *traspuesta*.
[Matriz traspuesta](https://upload.wikimedia.org/wikipedia/commons/e/e4/Matrix_transpose.gif). Este punto puede realizarlo fuera de las condiciones realizadas anteriormente.

**Extra**) puede controlar si un usuario introduce otro valor (letra, símbolo u otro caracter) usando `isdigit`, `is_numeric` o `is_decimal`. ¿Cuál vendría mejor? 


**En todo momento trate de usar listas de comprensión, siempre que le sea posible.**

````python
matrix = [[1, 2], [3, 4], [5, 6], [7, 8]]
numeros_mayores = []
# Parte 1
numero_usuario = int(input("Por favor ingrese un número: "))

# Parte 2
if(1 < numero_usuario < 9 ):
    for fila in range(len(matrix)):
      for columna in range(len(matrix[fila])):
        # Parte 3
        if matrix[fila][columna] > numero_usuario:
          numeros_mayores.append(matrix[fila][columna])

    print(numeros_mayores)

else:
  print("ERROR: El número no está entre 1 y 9")


# Parte 4
matrix_traspuesta = []
for indice in range(len(matrix[0])):
    fila_traspuesta = []

    for fila in matrix:
        fila_traspuesta.append(fila[indice])
    matrix_traspuesta.append(fila_traspuesta)

print("Matrix traspuesta usando bucles")
print(matrix_traspuesta)

matrix_traspuesta = [[fila[indice] for fila in matrix] for indice in range(2)]
print("Matrix traspuesta usando ")
print(matrix_traspuesta)
````
