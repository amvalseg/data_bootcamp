## Limpieza y preparaci√≥n de datos

```python
# Convertir 'null' strings a NaN reales
df = df.replace('null', np.nan)
```

```python
# Limpiar comillas en IDs
df['Booking ID'] = df['Booking ID'].str.replace('"', '')
df['Customer ID'] = df['Customer ID'].str.replace('"', '')
```

```python
# Convertir columnas num√©ricas
numeric_columns = ['Avg VTAT', 'Avg CTAT', 'Cancelled Rides by Customer', 
                   'Cancelled Rides by Driver', 'Incomplete Rides', 
                   'Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating']

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Crear columna datetime combinada
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
```

```python
# Extraer componentes temporales usando pandas
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['DayOfWeek'] = df['DateTime'].dt.day_name()
df['Hour'] = df['DateTime'].dt.hour
df['IsWeekend'] = df['DateTime'].dt.weekday >= 5
```

```python
# Crear categor√≠as de tiempo
def categorize_time(hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

df['TimeCategory'] = df['Hour'].apply(categorize_time)
```

```python
# Categorizar distancias usando NumPy
distances = df['Ride Distance'].dropna()
df['DistanceCategory'] = pd.cut(df['Ride Distance'], 
                               bins=np.percentile(distances, [0, 33, 66, 100]),
                               labels=['Short', 'Medium', 'Long'],
                               include_lowest=True)
```

```python
# Resultados del tratamiento
df[['DateTime', 'DayOfWeek', 'TimeCategory', 'DistanceCategory']].head()
```

### An√°lisis Bivariado y Multivariado

```python
# Matriz de correlaci√≥n para variables num√©ricas
numeric_cols = ['Avg VTAT', 'Avg CTAT', 'Booking Value', 'Ride Distance', 
                'Driver Ratings', 'Customer Rating']

correlation_data = df[numeric_cols].dropna()
if not correlation_data.empty:
    # Usando NumPy para calcular correlaciones
    correlation_matrix = np.corrcoef(correlation_data.T)
    correlation_df = pd.DataFrame(correlation_matrix, 
                                 columns=numeric_cols, 
                                 index=numeric_cols)
```

```python
correlation_df.round(3)
```

```python
# An√°lisis por tipo de veh√≠culo
vehicle_analysis = df.groupby('Vehicle Type').agg({
    'Booking Value': ['count', 'mean', 'sum'],
    'Ride Distance': 'mean',
    'Driver Ratings': 'mean',
    'Customer Rating': 'mean'
}).round(2)

vehicle_analysis
```

```python
# An√°lisis de conversi√≥n (tasa de √©xito)
conversion_rates = df.groupby('Vehicle Type')['Booking Status'].apply(
    lambda x: (x == 'Completed').mean() * 100
).round(2)
conversion_rates
```

```python
# AN√ÅLISIS DETALLADO DE CANCELACIONES
# Examinamos el impacto de las cancelaciones en el negocio
# Identificamos patrones y razones principales

# Filtrar solo reservas canceladas
cancellation_analysis = df[df['Booking Status'].str.contains('Cancel', na=False)]

# M√©tricas b√°sicas de cancelaci√≥n
total_cancellations = len(cancellation_analysis)
cancellation_rate = (total_cancellations / len(df) * 100)

# Crear resumen de cancelaciones
cancellation_summary = pd.Series({
    'Total_Cancellations': total_cancellations,
    'Cancellation_Rate_Percent': round(cancellation_rate, 2),
    'Total_Records': len(df)
})

cancellation_summary
```

```python
# RAZONES DE CANCELACI√ìN POR CLIENTE
# Analizamos por qu√© los clientes cancelan sus viajes
# Esto nos ayuda a identificar √°reas de mejora en el servicio

# Analizar razones de cancelaci√≥n por parte del cliente
customer_cancel_reasons = df['Reason for cancelling by Customer'].value_counts().dropna()

# Convertir a DataFrame con porcentajes
if len(customer_cancel_reasons) > 0:
    customer_cancellation_df = pd.DataFrame({
        'Count': customer_cancel_reasons,
        'Percentage': (customer_cancel_reasons / customer_cancel_reasons.sum() * 100).round(2)
    })
customer_cancellation_df
```

```python
# RAZONES DE CANCELACI√ìN POR CONDUCTOR
# Entendemos por qu√© los conductores cancelan viajes
# Esto puede indicar problemas operativos o de seguridad

# Analizar razones de cancelaci√≥n por parte del conductor
driver_cancel_reasons = df['Driver Cancellation Reason'].value_counts().dropna()

# Convertir a DataFrame con porcentajes
if len(driver_cancel_reasons) > 0:
    driver_cancellation_df = pd.DataFrame({
        'Count': driver_cancel_reasons,
        'Percentage': (driver_cancel_reasons / driver_cancel_reasons.sum() * 100).round(2)
    })
driver_cancellation_df
```

## An√°lisis Avanzado con Pandas

```python
# An√°lisis de comportamiento por mes
monthly_cohort = df.groupby(['Month', 'Booking Status']).size().unstack(fill_value=0)
monthly_cohort_pct = monthly_cohort.div(monthly_cohort.sum(axis=1), axis=0) * 100

monthly_cohort_pct.round(2)
```

```python
# An√°lisis RFM simplificado (Recency, Frequency, Monetary)
customer_metrics = df[df['Booking Status'] == 'Completed'].groupby('Customer ID').agg({
    'DateTime': lambda x: (df['DateTime'].max() - x.max()).days,  # Recency
    'Booking ID': 'count',  # Frequency
    'Booking Value': 'sum'   # Monetary
}).rename(columns={'DateTime': 'Recency', 'Booking ID': 'Frequency', 'Booking Value': 'Monetary'})

# Usar NumPy para crear percentiles y segmentos
customer_metrics['R_Score'] = pd.qcut(customer_metrics['Recency'], 5, labels=[5,4,3,2,1])
customer_metrics['F_Score'] = pd.qcut(customer_metrics['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
customer_metrics['M_Score'] = pd.qcut(customer_metrics['Monetary'], 5, labels=[1,2,3,4,5])

customer_metrics['RFM_Score'] = (customer_metrics['R_Score'].astype(str) + 
                                customer_metrics['F_Score'].astype(str) + 
                                customer_metrics['M_Score'].astype(str))

customer_metrics.head(10)
```

### Casos de Uso Pr√°cticos

#### Optimizaci√≥n de flota

```python
# An√°lisis de demanda por ubicaci√≥n y hora
demand_analysis = df.groupby(['Pickup Location', 'Hour']).agg({
    'Booking ID': 'count',
    'Booking Status': lambda x: (x == 'Completed').sum()
}).rename(columns={'Booking ID': 'Total_Demand', 'Booking Status': 'Successful_Rides'})

demand_analysis['Success_Rate'] = (demand_analysis['Successful_Rides'] / 
                                  demand_analysis['Total_Demand'] * 100).round(2)

top_demand = demand_analysis.groupby('Pickup Location')['Total_Demand'].sum().nlargest(10)
top_demand
```

### An√°lisis de rentabilidad

```python
# An√°lisis de rentabilidad por tipo de veh√≠culo
profitability = df[df['Booking Status'] == 'Completed'].groupby('Vehicle Type').agg({
    'Booking Value': ['sum', 'mean', 'count'],
    'Ride Distance': 'mean'
})

profitability.columns = ['Total_Revenue', 'Avg_Revenue', 'Trip_Count', 'Avg_Distance']
profitability['Revenue_per_km'] = (profitability['Total_Revenue'] / 
                                   (profitability['Avg_Distance'] * profitability['Trip_Count'])).round(2)

profitability.sort_values('Revenue_per_km', ascending=False)
```

### T√©cnicas Avanzadas

```python
# Tabla pivot para an√°lisis multidimensional
pivot_analysis = df.pivot_table(
    values='Booking Value',
    index='Vehicle Type',
    columns='TimeCategory',
    aggfunc=['mean', 'count'],
    fill_value=0
).round(2)

print("PIVOT: Valor promedio por veh√≠culo y horario")
pivot_analysis
```

```python
# Cross-tabulation
cross_tab = pd.crosstab(df['Vehicle Type'], df['Booking Status'], normalize='index') * 100
print("TABULACI√ìN CRUZADA: % de estados por veh√≠culo")
cross_tab.round(2)
```

```python
# Agrupaciones m√∫ltiples con funciones personalizadas
def efficiency_score(group):
    completed = (group['Booking Status'] == 'Completed').sum()
    total = len(group)
    avg_rating = group['Driver Ratings'].mean()
    if pd.isna(avg_rating):
        avg_rating = 0
    return completed / total * avg_rating

efficiency = df.groupby(['Vehicle Type', 'DayOfWeek']).apply(efficiency_score)
efficiency.nlargest(10).round(3)
```
